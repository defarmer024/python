```python
X = Data[['Smoking1 ','Belief_in_Afterlife1']]
X
```

```python
from sklearn.cluster import KMeans
```

```python
k_means=KMeans()
```

```python
k_means.fit(X)
```

```python
k_means.fit_predict(X)
```

```python
wcss=[]
for i in range(1,11):
    k_means = KMeans(n_clusters=i)
    k_means.fit(X)
    wcss.append(k_means.inertia_)
```

```python
wcss
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(range(1,11),wcss)
plt.title('Elbow Method')
plt.xlabel('Number OF Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
X = Data[['Smoking1 ','Belief_in_Afterlife1']]
k_means=KMeans(n_clusters=3,random_state=42)
y_means=k_means.fit_predict(X)
y_means
```

```python
plt.scatter(X.iloc[y_means==0,0],X.iloc[y_means==0,1],s=100,c='red',label='Cluster 1')
plt.scatter(X.iloc[y_means==1,0],X.iloc[y_means==1,1],s=100,c='blue',label='Cluster 2')
plt.scatter(X.iloc[y_means==2,0],X.iloc[y_means==2,1],s=100,c='yellow',label='Cluster 3')
plt.scatter(k_means.cluster_centers_[:,0], k_means.cluster_centers_[:,1],s=100,c='magenta')
plt.title('Belief In Afterlife And Smoking Segmentation')
plt.xlabel('Smoking')
plt.ylabel('Belief In AfterLife')
plt.legend()
plt.show()
```

```python
k_means.predict([[1,1]])
```
